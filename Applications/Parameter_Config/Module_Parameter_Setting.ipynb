# Example of how to update parameter and Package parameter objects

```python
obj = paru.Parameters({'hi':5})
obj2 = paru.Parameters({'outer':{'meow':10}})

obj.update(obj2)
print(obj)
```

```python
bo = paru.PackageParameters(
    data = {"myModule":{
        'hello':10
    }}
)

bo2 = paru.PackageParameters(
    data = {"myModule":{
        'hello2':20
    }}
)
bo.update(bo2)
print(bo)
    
    
```

```python
bo.module_attr_map(
    attr_list = ["hell2"],
    module_name = "myModule",
    plus_unused=True
)
```
